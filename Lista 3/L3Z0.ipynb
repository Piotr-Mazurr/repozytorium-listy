```python
a=float(input("Podaj liczbę a:"))
b=float(input("Podaj liczbę b:"))
c=float(input("Podaj liczbę c:"))
d=float(input("Podaj liczbę d:"))
sequence=[a,b,c,d]
s=0
k=0
for x in sequence:
    s=s+x
    k=k+1
    print(f"{k} iteracja: s = {s}")


#iteracja 1: do s=0 dodawana jest pierwsza pozycja z listy tzn. x=a: na wyjściu mamy s = s + x = 0 + a = a
#iteracja 2: do s = a dodawana jest druga pozycja z listy tzn. x=b: na wyjściu mamy s = s + x = a + b
#itearcja 3: do s = a + b, dodawana jest trzecia pozycja z listy tzn. x=c: na wyjściu mamy s = s + x = a + b + c
#iteracja 4: do s = a + b + c dodawana jest czwarta pozycja z listy tzn. x=d: na wyjściu mamy s = s + x = a + b + c + d
```

```python
n=int(input("Ile wyrazów ciągu chcesz zsumować?:"))
sequence=[n**n + (n+3)**2 for n in range (0,n+1)]
#sequence to lista wyrazów ciagu, indeksujac od 0

print("kolejne wyrazy cigu")
k=0
for x in sequence:
    print(f"a_{k} = {x}")
    k=k+1

print("kolejne sumy wyrazów ciagu")
s=0
k=0
for x in sequence:
    s=s+x
    print(f"S_{k} = {s}")
    k=k+1
```
