```python
import matplotlib.pyplot as plt
from math import sin, cos, pi, tan
from matplotlib.pyplot import grid
```

```python
#podpunkt a
xn = [(2*pi * k ) / 100 for k in range (0, 101)]

x = [cos(t) for t in xn]
y = [sin(t) for t in xn]
plt.plot(x,y)
grid()
plt.show()
```

```python
#podpunkt b
xn = [(2*pi * k ) / 100 for k in range (0, 101)]

x = [cos(t) for t in xn[0:26]]
y = [sin(t) for t in xn[0:26]]
plt.plot(x,y, color="red")

x = [cos(t) for t in xn[25:51]]
y = [sin(t) for t in xn[25:51]]
plt.plot(x,y, color="blue")

x = [cos(t) for t in xn[50:76]]
y = [sin(t) for t in xn[50:76]]
plt.plot(x,y, color="yellow")

x = [cos(t) for t in xn[75:101]]
y = [sin(t) for t in xn[75:101]]
plt.plot(x,y, color="green")

grid()
plt.show()
```

```python
#podpunkt c
xn = [(2*pi * k ) / 100 for k in range (0, 101)]

x = [cos(t+pi/4) for t in xn[0:26]]
y = [1/2*sin(t+pi/4) for t in xn[0:26]]
plt.plot(x,y, color="red")

x = [cos(t+pi/4) for t in xn[25:51]]
y = [1/2*sin(t+pi/4) for t in xn[25:51]]
plt.plot(x,y, color="blue")

x = [cos(t+pi/4) for t in xn[50:76]]
y = [1/2*sin(t+pi/4) for t in xn[50:76]]
plt.plot(x,y, color="yellow")

x = [cos(t+pi/4) for t in xn[75:101]]
y = [1/2*sin(t+pi/4) for t in xn[75:101]]
plt.plot(x,y, color="green")

grid()
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.show()
```

```python
#zadanie 2 metodami wykadowymi
x0=1
y0= 0

x = [x0]
y = [y0]

for _ in range(500):
  nowe_x0 = 22220/22100 * x0 + 2121/22100 * y0
  nowe_y0 = -2121/22100 * x0 + 22220/22100 * y0

  x.append(nowe_x0)
  y.append(nowe_y0)

  x0 = nowe_x0
  y0 = nowe_y0


plt.plot(x,y, color="red")
grid()
plt.show()
```

```python
n=10 #ilo okresw tangensa na wykresie
tn = [pi*k/100 for k in range (-49,50)]


#rysowanie tangensa
k=0
for _ in range(n):
  x1=[p + pi*k for p in tn]
  k=k+1

  x=[t for t in x1]
  y=[tan(t) for t in tn]

  plt.plot(x, y, color="green")


#rysowanie asymptot
k=0
for _ in range(n+1):
  plt.axvline(x=-pi/2+pi*k)
  k=k+1


grid()
plt.show()
```
