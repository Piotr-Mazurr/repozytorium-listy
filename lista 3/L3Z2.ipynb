```python
an=[1,1,1]

def moving_sum(an):
    #tworzymy listę p odpowiedniej długości (1 element krótsza niż ilość wyrazów ciągu na)
    p = [0 for k in range(1,len(an))]

    #zastępujemy elementy listy p
    for i in range (0,len(an)-1):
        p[i] = an[i] + an[i+1]

    return p

print(moving_sum(an))
```
